```python
# DOCUMENTAÇÃO DA API
# https://github.com/martin-majlis/Wikipedia-API


lingua_origem = "pt"

instancias_wikipedias_linguas = {}

instancias_wikipedias_linguas[lingua_origem] = wikipediaapi.Wikipedia(lingua_origem)

dicionario_artigos = {}
```

```python
      



def percorrer_linguas (pagina_origem, profundidade=0):
    
    chave_pagina_origem = f'{pagina_origem.language}-{pagina_origem.title}'  
    
    #O adiciona no dicionário de artigos
    dicionario_artigos[chave_pagina_origem] = {
                                                'pagina': {
                                                    'titulo':pagina_origem.title,
                                                    'lingua':pagina_origem.language,
                                                    'texto':pagina_origem.text
                                                }
                                              }                                                
            
    
    links_linguas = pagina_origem.langlinks
    
    #Percorre os links para este artigo em outras línugas
    for chave_link_lingua in sorted(links_linguas.keys()):
        
            
        link_lingua = links_linguas[chave_link_lingua]
        
        
        print(f'{"---"*profundidade}{chave_link_lingua}: {link_lingua.language}, - {link_lingua.title}: {link_lingua.fullurl}')
        
        
        #Verifica se ja foi instanciada a Wikipedia esta língua
        if not link_lingua.language in instancias_wikipedias_linguas:
            
            #Se não existir cria e adiciona ao nosso dicionário de instâncias da Wikipedia
            instancias_wikipedias_linguas[link_lingua.language] = wikipediaapi.Wikipedia(link_lingua.language)
            
            
            
        #A chave da página é a lingua e o título
        chave_pagina_outra_lingua = f'{link_lingua.language}-{link_lingua.title}'    
        
        
        #Se o artigo ainda não existe para essa lingua no dicionário
        if not chave_pagina_outra_lingua in dicionario_artigos:
            
            #Puxa a página da wikipedia na outra língua
            pagina_outra_lingua = instancias_wikipedias_linguas[link_lingua.language].page(link_lingua.title)
            
            #Percorre as línguas deste artigo
            percorrer_linguas(pagina_outra_lingua, profundidade + 1)
            
        
        
        
```

```python
titulo_pagina_origem = 'Hoʻoponopono'
```

```python
import json
import os
import wikipediaapi

nome_arquivo = f'dicionario_artigos_{titulo_pagina_origem}.json'

if os.path.exists(nome_arquivo):
    
    with open(nome_arquivo, 'r') as arquivo_entrada:
        
        dicionario_artigos = json.load(arquivo_entrada)
    

else:
    
    percorrer_linguas(instancias_wikipedias_linguas[lingua_origem].page(titulo_pagina_origem))

 
    #Salva o arquivo de artigos
    with open(nome_arquivo, "w")  as arquivo_saida:
    
        arquivo_saida.write(json.dumps(dicionario_artigos, indent=4))
```
