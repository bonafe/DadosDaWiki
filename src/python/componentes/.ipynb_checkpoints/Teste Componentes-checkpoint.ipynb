```python
df_credits = pd.read_csv( os.path.join(pasta_dados,"credits.csv"), index_col=None, header=0)
df_credits.name = "credits"

df_keywords = pd.read_csv( os.path.join(pasta_dados,"keywords.csv"), index_col=None, header=0)
df_keywords.name = "keywords"

df_links = pd.read_csv( os.path.join(pasta_dados,"links.csv"), index_col=None, header=0)
df_links.name = "links"

df_links_small = pd.read_csv( os.path.join(pasta_dados,"links_small.csv"), index_col=None, header=0)
df_links_small.name = "links_small"

df_movies_metadata = pd.read_csv( os.path.join(pasta_dados,"movies_metadata.csv"), index_col=None, header=0)
df_movies_metadata.name = "movies_metadata"

df_ratings = pd.read_csv( os.path.join(pasta_dados,"ratings.csv"), index_col=None, header=0)
df_ratings.name = "ratings"

df_ratings_small = pd.read_csv( os.path.join(pasta_dados,"ratings_small.csv"), index_col=None, header=0)
df_ratings_small.name = "ratings_small"
```

```python
df_movies_metadata = pd.read_csv( os.path.join(pasta_dados,"movies_metadata.csv"), index_col=None, header=0)
df_movies_metadata.name = "movies_metadata"
```

```python
dfs_originais = [
    df_credits, 
    df_keywords, 
    df_links, 
    df_links_small, 
    df_movies_metadata,  
    df_ratings, 
    df_ratings_small
]

dfs = []

for df_original in dfs_originais:
    
    df = None
    
    try:
        df = df_original.copy()
    except:
        df = df_original.copy()
        
    if hasattr(df_original,"name"):
        df.name = df_original.name        
    
    dfs.append(df)
```
